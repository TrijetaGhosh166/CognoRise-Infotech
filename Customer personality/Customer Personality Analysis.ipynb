```python
# Creating a dictionary for manual label encoding
e_mapping = {
    " Graduate": 0,
    "Postgraduate": 1,
    "Undergraduate": 2
}
df2['Education']=df2['Education'].map(e_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    df2[i]=df2[[i]].apply(LE.fit_transform)
```

```python
print(df2["Relationship"].value_counts())
print(df2["Education"].value_counts())
```

```python
#Creating a copy of data
df3=df2.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4','AcceptedCmp5', 'Complain', 'Response', 'Age']
df3=df3.drop(cols, axis=1)
#Scaling
scaler=StandardScaler()
scaler.fit(df3)
scaled=pd.DataFrame(scaler.transform(df3),columns= df3.columns)
print("All features are now scaled")
```

```python
scaled.head(5)
```

### Dimension reduce

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled)
PCA_ds = pd.DataFrame(pca.transform(scaled), columns=["col1", "col2", "col3"])
PCA_ds.describe().T

```

### Clustering

```python
from sklearn.cluster import AgglomerativeClustering
# Create an instance of AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
# Fit the model to the scaled data
clusters = agg_clustering.fit_predict(PCA_ds)
# Add the cluster labels to the PCA dataset
PCA_ds['Cluster'] = clusters
# View the clusters
print(PCA_ds['Cluster'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Scatter plot of the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(PCA_ds['col1'], PCA_ds['col2'], c=PCA_ds['Cluster'], cmap='viridis', s=50)
plt.title('Hierarchical Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()

```
